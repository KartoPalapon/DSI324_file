```python
import os
import json
import pandas as pd

# โฟลเดอร์ที่เก็บไฟล์ JSON
folder_path = r"C:\Users\U s e r\OneDrive\Desktop\DSI324_file\data\3_output\basic_info"

# รายการข้อมูลที่จะแตกออกเป็น record
rows = []

# วนลูปทุกไฟล์ในโฟลเดอร์
for filename in os.listdir(folder_path):
    if filename.endswith(".json"):
        file_path = os.path.join(folder_path, filename)
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                content = json.load(f)

                # ดึงข้อมูลหลัก พร้อม fallback เป็น None ถ้าไม่มี
                
                cur_name_th = content.get("cur-name-th", None)
                faculty_name_th = content.get("faculty_name_th", None)
                careers = content.get("careers", None)

                # ถ้ามี careers แบบ list และไม่ว่าง → แตกแถว
                if isinstance(careers, list) and len(careers) > 0:
                    for career in careers:
                        rows.append([
        
                            cur_name_th,
                            faculty_name_th,
                            career.strip() if isinstance(career, str) else None
                        ])
                else:
                    # ถ้าไม่มี careers หรือเป็น null → สร้างแถวเดียว
                    rows.append([
                        
                        cur_name_th,
                        faculty_name_th,
                        None
                    ])
        except Exception as e:
            print(f"❌ Error reading {filename}: {e}")

# บังคับ schema ชัดเจน
columns = [ "cur_name_th", "faculty_name_th", "career"]
df = pd.DataFrame(rows, columns=columns)

# แสดงผลตัวอย่าง
print(df.head())

# บันทึกไฟล์
df.to_csv("careers_extracted.csv", index=False, encoding="utf-8-sig")
```

```python
import os
import json
import pandas as pd

# โฟลเดอร์ที่เก็บไฟล์ JSON
folder_path = r"C:\Users\U s e r\OneDrive\Desktop\DSI324_file\data\3_output\final"

# รายการข้อมูลที่จะเก็บในตาราง
rows = []

# รายชื่อคอลัมน์ที่ต้องการ (ระบุชัดเจน)
columns = [
    "cur-name-th", "faculty_name_th", "course_code", "th-abbr", "course_name_th", "course_name_en",
    "course_type", "course_credit", "lect_hours", "practice_hours",
    "self_study_hours", "course_description_th", "course_description_en", "prerequisite_courses"
]

# วนลูปอ่านไฟล์ JSON
for filename in os.listdir(folder_path):
    if filename.endswith(".json"):
        file_path = os.path.join(folder_path, filename)
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                content = json.load(f)
                courses = content.get("courses", [])

                # ดึงข้อมูลที่อยู่นอก course object
                cur_name_th = content.get("cur-name-th", None)
                faculty_name_th = content.get("faculty_name_th", None)

                # ถ้ามี course list
                if isinstance(courses, list):
                    for course in courses:
                        row = []
                        for col in columns:
                            # ถ้าเป็นฟิลด์ที่อยู่ระดับนอก (ใน content) ให้ดึงจาก content
                            if col == "cur-name-th":
                                row.append(cur_name_th)
                            elif col == "faculty_name_th":
                                row.append(faculty_name_th)
                            else:
                                row.append(course.get(col, None))
                        rows.append(row)
                else:
                    print(f"⚠️ No course list in {filename}")

        except Exception as e:
            print(f"❌ Error reading {filename}: {e}")

# สร้าง DataFrame โดยใช้ columns ที่ระบุไว้
df = pd.DataFrame(rows, columns=columns)

# แสดงตัวอย่าง
print(df.head())

# บันทึกเป็นไฟล์ CSV
df.to_csv("courses_extracted.csv", index=False, encoding="utf-8-sig")


```

```python
import os
import json
import pandas as pd

# โฟลเดอร์ที่เก็บไฟล์ JSON
folder_path = r"C:\Users\U s e r\OneDrive\Desktop\DSI324_file\data\3_output\final"

# รายการข้อมูลที่จะเก็บในตาราง
rows = []

# รายชื่อคอลัมน์ที่ต้องการ
columns = [
    "cur-name-th", "faculty_name_th",
    "PLO_Type", "PLO_detail"
]

# วนลูปอ่านไฟล์ JSON
for filename in os.listdir(folder_path):
    if filename.endswith(".json"):
        file_path = os.path.join(folder_path, filename)
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                content = json.load(f)

                # ดึงข้อมูลที่อยู่นอก course object
                cur_name_th = content.get("cur-name-th", None)
                faculty_name_th = content.get("faculty_name_th", None)

                # ดึง PLOs ถ้ามี
                plos = content.get("ผลลัพธ์การเรียนรู้ของหลักสูตร (PLOs)", {})

                for plo_type, details in plos.items():
                    for detail in details:
                        row = [
                            cur_name_th,
                            faculty_name_th,
                            plo_type,
                            detail
                        ]
                        rows.append(row)

        except Exception as e:
            print(f"❌ Error reading {filename}: {e}")

# สร้าง DataFrame โดยใช้ columns ที่ระบุไว้
df = pd.DataFrame(rows, columns=columns)

# แสดงตัวอย่าง
print(df.head())

# บันทึกเป็นไฟล์ CSV
df.to_csv("PLOs_extracted.csv", index=False, encoding="utf-8-sig")
```

```python
import os
import json
import pandas as pd

# โฟลเดอร์ที่เก็บไฟล์ JSON
folder_path = r"C:\Users\U s e r\OneDrive\Desktop\DSI324_file\data\3_output\final"

# รายการข้อมูลที่จะเก็บในตาราง
rows = []

# กำหนด columns ที่ต้องการ
columns = [
    "faculty_name_th", "faculty_name_en", "curr-id",
    "cur-name-th", "cur-name-en",
    "degree-full-th", "degree-full-en", "degree-abr-th", "degree-abr-en",
    "curr-type", "lang", "campus",
    "approve-date", "first-open-term",
    "expense-type", "student-nation-id",
    "admit-plan-yearly", "max-years",
    "total-credits", "gen-ed-credits", "req-credits", "elec-credits",
    "intern-credits", "coop-credits"
]

# วนลูปอ่านไฟล์ JSON
for filename in os.listdir(folder_path):
    if filename.endswith(".json"):
        file_path = os.path.join(folder_path, filename)
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                content = json.load(f)

                # ดึงค่าตาม columns ที่กำหนด
                row = []
                for col in columns:
                    row.append(content.get(col, None))
                rows.append(row)

        except Exception as e:
            print(f"❌ Error reading {filename}: {e}")

# สร้าง DataFrame
df = pd.DataFrame(rows, columns=columns)

# แสดงตัวอย่าง
print(df.head())

# บันทึกเป็นไฟล์ CSV
df.to_csv("curriculum_metadata.csv", index=False, encoding="utf-8-sig")
```
