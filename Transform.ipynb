```python
import pandas as pd
import os
import glob
import json

# 1. โหลดข้อมูลจาก CSV
csv_path = r'C:\Users\U s e r\OneDrive\Desktop\DSI324_file\curriculum_metadata.csv'
df = pd.read_csv(csv_path)

# ตรวจสอบว่ามีคอลัมน์ที่ต้องใช้
if 'curr-id' not in df.columns or 'cur-name-th' not in df.columns:
    raise Exception("❌ CSV ต้องมีทั้ง 'curr-id' และ 'cur-name-th'")

# สร้าง dictionary: {cur-name-th: curr-id}
cur_name_to_id = dict(zip(df['cur-name-th'], df['curr-id'].astype(str)))

# 2. ค้นหาไฟล์ JSON
folder_path = r'C:\Users\U s e r\OneDrive\Desktop\DSI324_file\data\3_output\final'
json_files = glob.glob(os.path.join(folder_path, '*.json'))

updated = 0
skipped = 0

# 3. อ่านและอัปเดตไฟล์ JSON ทีละไฟล์
for file_path in json_files:
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)

        cur_name = data.get('cur-name-th', None)

        if cur_name and cur_name in cur_name_to_id:
            data['curr-id'] = cur_name_to_id[cur_name]
            updated += 1

            with open(file_path, 'w', encoding='utf-8') as f:
                json.dump(data, f, ensure_ascii=False, indent=4)
        else:
            skipped += 1

    except Exception as e:
        print(f"⚠️ Error processing file {file_path}: {e}")
        skipped += 1

# 4. รายงานผล
print(f"✅ อัปเดตแล้ว {updated} ไฟล์")
print(f"⏭️ ข้าม {skipped} ไฟล์ (ไม่พบ 'cur-name-th' หรือตรงกับ CSV ไม่ได้)")
```

```python
import pandas as pd
import os
import glob
import json

# โฟลเดอร์ที่เก็บไฟล์ JSON
folder_path = r'C:\Users\U s e r\OneDrive\Desktop\DSI324_file\data\3_output\final'
json_files = glob.glob(os.path.join(folder_path, '*.json'))

# สร้างลิสต์เก็บข้อมูล
records = []

for file in json_files:
    try:
        with open(file, 'r', encoding='utf-8') as f:
            data = json.load(f)

            # ดึง curr-id และ institution (แม้เป็นลิสต์ว่าง)
            curr_id = data.get('curr-id', None)
            institution = data.get('institution', [])

            # แปลง institution list ให้เป็นข้อความเดียว (ถ้าไม่ว่าง)
            institution_str = '; '.join(institution) if institution else None

            records.append({'curr-id': curr_id, 'institution': institution_str})
    except Exception as e:
        print(f"⚠️ Error in file {file}: {e}")

# แปลงเป็น DataFrame และบันทึก CSV
df = pd.DataFrame(records)
df.to_csv('collaboration.csv', index=False, encoding='utf-8-sig')
print("✅ Saved all data to collaboration.csv")
```

```python
import os
import json
import pandas as pd

# โฟลเดอร์ที่เก็บไฟล์ JSON
folder_path = r"C:\Users\U s e r\OneDrive\Desktop\DSI324_file\data\3_output\final"

# รายการข้อมูลที่จะเก็บในตาราง
rows = []

# รายชื่อคอลัมน์ที่ต้องการ (ระบุชัดเจน)
columns = [
    "curr-id", "course_code", "th-abbr", "course_name_th", "course_name_en",
    "course_type", "course_credit", "lect_hours", "practice_hours",
    "self_study_hours", "course_description_th", "course_description_en", "prerequisite_courses"
]

# วนลูปอ่านไฟล์ JSON
for filename in os.listdir(folder_path):
    if filename.endswith(".json"):
        file_path = os.path.join(folder_path, filename)
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                content = json.load(f)
                courses = content.get("courses", [])

                # ดึงข้อมูลที่อยู่นอก course object
                curr_id = content.get("curr-id", None) # เปลี่ยนตัวแปรเป็น curr_id

                # ถ้ามี course list
                if isinstance(courses, list):
                    for course in courses:
                        row = []
                        for col in columns:
                            # เพิ่มเงื่อนไขตรวจสอบคอลัมน์เพื่อดึงข้อมูลจากตำแหน่งที่ถูกต้อง
                            row.append(course.get(col, None))
                        rows.append(row)
                else:
                    print(f"⚠️ No course list in {filename}")

        except Exception as e:
            print(f"❌ Error reading {filename}: {e}")

# สร้าง DataFrame โดยใช้ columns ที่ระบุไว้
df = pd.DataFrame(rows, columns=columns)

# แสดงตัวอย่าง
print(df.head())

# บันทึกเป็นไฟล์ CSV
df.to_csv("courses_extracted.csv", index=False, encoding="utf-8-sig")
```

```python
import os
import glob
import json
import pandas as pd

folder_path = r'C:\Users\U s e r\OneDrive\Desktop\DSI324_file\data\3_output\final'
json_files = glob.glob(os.path.join(folder_path, '*.json'))

records = []

for file in json_files:
    try:
        with open(file, 'r', encoding='utf-8') as f:
            data = json.load(f)
            curr_id = data.get('curr-id', None)
            courses = data.get('courses', [])

            for course in courses:
                records.append({
                    'curr-id': curr_id,
                    'course_code': course.get('course_code'),
                    'th-abbr': course.get('th-abbr'),
                    'course_name_th': course.get('course_name_th'),
                    'course_name_en': course.get('course_name_en'),
                    'course_type': course.get('course_type'),
                    'course_credit': course.get('course_credit'),
                    'lect_hours': course.get('lect_hours'),
                    'practice_hours': course.get('practice_hours'),
                    'self_study_hours': course.get('self_study_hours'),
                    'course_description_th': course.get('course_description_th'),
                    'course_description_en': course.get('course_description_en'),
                    'prerequisite_courses': course.get('prerequisite_courses')
                })

    except Exception as e:
        print(f"⚠️ Error reading {file}: {e}")

# แปลงเป็น DataFrame
df = pd.DataFrame(records)

# แปลง prerequisite_courses ให้เป็นข้อความ (ถ้าต้องการ)
df['prerequisite_courses'] = df['prerequisite_courses'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)

# บันทึกเป็น CSV
df.to_csv('course_details.csv', index=False, encoding='utf-8-sig')
print("✅ Exported to course_details.csv")
```

```python
import os
import glob
import json
import pandas as pd

# ระบุโฟลเดอร์ที่เก็บไฟล์ JSON
folder_path = r'C:\Users\U s e r\OneDrive\Desktop\DSI324_file\data\3_output\final'
json_files = glob.glob(os.path.join(folder_path, '*.json'))

records = []

for file in json_files:
    try:
        with open(file, 'r', encoding='utf-8') as f:
            data = json.load(f)
            curr_id = data.get('curr-id', None)
            professors = data.get('professor', [])

            for prof in professors:
                records.append({
                    'curr-id': curr_id,
                    'lecture-position': prof.get('lecture-position'),
                    'lecture-first-name': prof.get('lecture-first-name'),
                    'lecture-last-name': prof.get('lecture-last-name'),
                    'bachelor-degree': prof.get('bachelor-degree'),
                    'bachelor-institute': prof.get('bachelor-institute'),
                    'bachelor-graduate-year': prof.get('bachelor-graduate-year'),
                    'doctor-degree': prof.get('doctor-degree'),
                    'doctor-institute': prof.get('doctor-institute'),
                    'doctor-graduate-year': prof.get('doctor-graduate-year')
                })

    except Exception as e:
        print(f"⚠️ Error reading {file}: {e}")

# สร้าง DataFrame
df = pd.DataFrame(records)

# บันทึกเป็น CSV
df.to_csv('professor_details.csv', index=False, encoding='utf-8-sig')
print("✅ Exported to professor_details.csv")
```

```python
import pandas as pd
import random

# โหลดไฟล์
professor_df = pd.read_csv('professor.csv')
curriculum_df = pd.read_csv('curriculum_metadata.csv')

# รวมข้อมูลด้วย curr-id
merged_df = professor_df.merge(curriculum_df[['curr-id', 'faculty_id']], on='curr-id', how='left')

# เตรียมตัวนับสำหรับกรณี faculty_id หาย
missing_faculty_counter = 0

# สำหรับเก็บรหัสที่ใช้ไปแล้ว เพื่อหลีกเลี่ยงการซ้ำ
used_ids = {}

# ฟังก์ชันสร้าง professor_id
def generate_professor_id(row):
    global missing_faculty_counter
    faculty_id = row['faculty_id']
    
    if pd.isna(faculty_id):
        pid = f"{missing_faculty_counter:04}"
        missing_faculty_counter += 1
        return pid
    else:
        faculty_id_str = f"{int(faculty_id):02}"
        # ตรวจสอบว่าใช้เลขสุ่มไปแล้วหรือยัง
        if faculty_id_str not in used_ids:
            used_ids[faculty_id_str] = set()

        # หาเลขสุ่มที่ยังไม่เคยใช้
        while True:
            random_part = f"{random.randint(0, 99):02}"
            pid = faculty_id_str + random_part
            if random_part not in used_ids[faculty_id_str]:
                used_ids[faculty_id_str].add(random_part)
                return pid

# สร้าง professor_id
merged_df['professor_id'] = merged_df.apply(generate_professor_id, axis=1)

# ย้าย professor_id ไปเป็นคอลัมน์แรก
cols = ['professor_id'] + [col for col in merged_df.columns if col != 'professor_id']
final_df = merged_df[cols]

# บันทึกไฟล์
final_df.to_csv('professor_with_id.csv', index=False, encoding='utf-8-sig')
print("✅ Saved to professor_with_id.csv")
```
